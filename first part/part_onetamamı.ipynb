## Merging 5 datasets into a dataset

### Creating empty dataset

```python
new_frame = pd.DataFrame(np.zeros([155,10]),columns = df1.columns)
new_frame.head()
```

### Filling up all columns 

```python
countries = merged[2015]["Country"]
new_frame["Country"] = countries
columns = list(merged[2015].columns)[3:len(list(merged[2015].columns))]

for i in range(len(new_frame.Country.values)):
    index_2015 = merged[2015].loc[merged[2015]["Country"] == new_frame.Country.values[i]].index
    index_2016 = merged[2016].loc[merged[2016]["Country"] == new_frame.Country.values[i]].index
    index_2017 = merged[2017].loc[merged[2017]["Country"] == new_frame.Country.values[i]].index
    index_2018 = merged[2018].loc[merged[2018]["Country"] == new_frame.Country.values[i]].index
    index_2019 = merged[2019].loc[merged[2019]["Country"] == new_frame.Country.values[i]].index
    for x in columns:
        counter = 0
        summ = 0
        if len(merged[2015][x][index_2015].values) != 0:
            summ += merged[2015][x][index_2015].values[0]
            counter += 1
            
        if len(merged[2016][x][index_2016].values) != 0:
            summ += merged[2016][x][index_2016].values[0]
            counter += 1
        
        if len(merged[2017][x][index_2017].values) != 0:
            summ += merged[2017][x][index_2017].values[0]
            counter += 1
            
        if len(merged[2018][x][index_2018].values) != 0:
            summ += merged[2018][x][index_2018].values[0]
            counter += 1
        
        if len(merged[2019][x][index_2019].values) != 0:
            summ += merged[2019][x][index_2019].values[0]
            counter += 1
        
        new_frame[x][i] = summ / counter
```
